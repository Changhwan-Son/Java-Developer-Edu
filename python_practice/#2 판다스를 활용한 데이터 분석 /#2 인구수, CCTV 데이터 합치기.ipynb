# 상관계수
x 축 : 마신 술 <br/>
y 축 : 혈중 알콜 농도 <br/>
직선의 기울기 : 상관 계수 <br/>

어떠한 데이터끼리 변화에 대한 상관관계를 구한 결과를 상관계수라고 한다. 

* x가 증가했을 때 y가 얼마나 증가 / 감소 했는가 
* 상관계수의 절댓값이 0.1 이하면 관계가 거의 없는 상관관계 
* 상관계수의 절댓값이 0.3 이하면 약한 상관관계 
* 상관계수의 절댓값이 0.7 이하면 뚜렷한 상관관계 

-> 피어슨 상관계수

인구수 대비 CCTV 대수 확인 

```python
plt.figure(figsize=(10,10))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV 소계')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)
# 인구수를 x로 놓고, y를 소계로 놓는 1차 함수 ([기울기, 절편])
fp1
```

```python
f1 = np.poly1d(fp1) # fp1의 정보를 이용해서 ( 기울기, 절편 ) 1차 함수 수식을 만들어 낸다.

# 함수에 들어갈 값을 만든다. 
fx = np.linspace(100000, 700000, 100)  # 10만부터 70만까지 100의 간격으로 정수 배열을 만들어 낸다. 

plt.figure(figsize=(10,10))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)

# 상관 계수를 활용한 직선 긋기 
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV 소계')
plt.grid()
plt.show()
```

대략 300,000명의 인구수 일 때 1,100개의 CCTV를 설치하는 것이 기준이 될 수도 있겠구나. 

* 직선보다 위에 있으면 인구수에 비해 CCTV가 많이 설치된 것 
* 직선보다 아래에 있으면 인구수에 비해 CCTV가 적게 설치된 것

# 마지막 정리하기
인구수에 비해 CCTV가 비정상적으로 많거나 적은 구가 어딘지를 보기 위해 

1. 직선과 실제 데이터와의 오차 구하기 
2. 오차의 크기로 정렬
3. 다시 시각화 

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

# 오차 구하기 
data_result['오차'] = np.abs(data_result['소계'] - f1(data_result['인구수']))

# 오차별로 정렬
df_sort = data_result.sort_values(by ='오차', ascending=False)
df_sort.head()
```

구의 이름을 텍스트로 같이 표시

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'], data_result['소계'], c=data_result['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g') #x축 기준 인구수, y축 기준 CCTV 대수

#오차가 가장 큰 10개의 구에 대해서 텍스트 부여하기
for n in range(10):
    #scatter가 찍힌 좌표에 적당히 거리를 조절하여 해당하는 데이터의 인덱스(df_sort.index[n])를 출력하기
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n] * 0.98, df_sort.index[n], fontsize=15)
    
plt.xlabel('인구수')
plt.ylabel('인구당비율')

plt.colorbar()
plt.grid()
plt.show()
```
